```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(88)

# x=np.linspace(start,stop,num) , 개수 기준으로 배열 생성
df = pd.DataFrame({'A':np.linspace(1, 10, 10)})
print(df)


# 10 * 4 배열 생성
# pd.concat() 데이터 연결 함수
# pd.concat([s1,s2])

df = pd.concat([df, pd.DataFrame(np.random.randn(10,4), columns = list('BCDE'))],
               axis =1)

# 널값 입력
df.iloc[3,3] = np.nan
df.iloc[0,2] = np.nan
```
